```python
import numpy as np

# Input
X = np.array([[0,0],[0,1],[1,0],[1,1]])

print("XOR Gate (Hardcoded Weights)")

# Step function
def step(x):
    return 1 if x > 0 else 0

# Hidden layer (2 neurons)
def hidden_layer(x):
    h1 = step(x[0] + x[1] - 0.5)   # OR
    h2 = step(x[0] + x[1] - 1.5)   # AND
    return np.array([h1, h2])

# Output layer
def output_layer(h):
    return step(h[0] - 2*h[1] - 0.5)

# Predictions
for x in X:
    h = hidden_layer(x)
    y = output_layer(h)
    print(f"{x} → {y}")
```
